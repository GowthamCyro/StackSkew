```python
from scipy.stats import norm

# (i) P(X >= 20)
z1 = (20-12)/4
prob_1 = 1 - norm.cdf(z1)
print("Probability of X being greater than or equal to 20:", prob_1.round(2))

# (ii) P(X <= 20)
prob_2 = norm.cdf(z1)
print("Probability of X being less than or equal to 20:", prob_2.round(2))

# (iii) P(0 <= X <= 12)
z2 = (12-12)/4
z3 = (0-12)/4
prob_3 = norm.cdf(z2) - norm.cdf(z3)
print("Probability of X being between 0 and 12 (inclusive):", prob_3.round(2))
```

**2nd Question**

```python
import scipy.stats as stats

mean = 662
sd = 32
n = 1000

# (i) Proportion of plots with yield over 700 kilos
z1 = (700 - mean) / sd
prop_over_700 = 1 - stats.norm.cdf(z1)     #cumulative distribution function
plots_over_700 = round(prop_over_700 * n)
print(f"Number of plots with yield over 700 kilos: {plots_over_700}")

# (ii) Proportion of plots with yield below 650 kilos
z2 = (650 - mean) / sd     
prop_below_650 = stats.norm.cdf(z2)     #cumulative distribution function
plots_below_650 = round(prop_below_650 * n)
print(f"Number of plots with yield below 650 kilos: {plots_below_650}")

# (iii) Lowest yield of the best 100 plots
z3 = stats.norm.ppf(0.01)    # percent point function
lowest_yield = round(mean + z3 * sd, 2)
print(f"Lowest yield of the best 100 plots: {lowest_yield}")
```

**4th Question**

```python
from scipy.stats import poisson

mean_demand = 2

# (i) Probability that neither car is used
prob_no_demands = poisson.pmf(0, mean_demand)
prob_neither_car_used = prob_no_demands ** 5  #power 5 beacuse we have 5 ways to choose.
print(f"Proportion of days on which neither car is used: {prob_neither_car_used:.4f}")

# (ii) Probability that two demands are there
prob_two_demands = poisson.pmf(2, mean_demand)
prob_two_demands_all_unused = prob_two_demands * (5 * 4 / 2) * prob_no_demands ** 3 #power 3 beacuse we have 3 ways to choose.
print(f"Proportion of days on which two demands occur and all five cars are not used: {prob_two_demands_all_unused:.4f}")

# (iii) Probability that all cars are used
prob_all_used = poisson.pmf(5, mean_demand)
print(f"Proportion of days on which all cars are used: {prob_all_used:.4f}")
```

**5th question**

```python
import numpy as np
from scipy.stats import norm

mu = 345
sigma = np.sqrt(6)
n = 25
x_bar = 348

# Calculate the test statistic
z = (x_bar - mu) / (sigma / np.sqrt(n))

# Calculate the p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Print the results
print(f"Test statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("The machine is malfunctioning.")
else:
    print("The machine is not malfunctioning.")
```

**6th question**

```python
import numpy as np
from scipy.stats import chisquare

observed = np.array([[120, 80], [130, 70], [70, 30], [80, 20]])

# Calculate the expected values under the null hypothesis of independence
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
grand_total = observed.sum()
expected = np.outer(row_totals, col_totals) / grand_total

chi2, p = chisquare(observed.flatten(), expected.flatten(), ddof=2)

print("Chi-square test statistic:", chi2)
print("P-value:", p)
```
